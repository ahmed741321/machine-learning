```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df
```

```python
df = pd.get_dummies(df, columns=['Sex'], drop_first=True,dtype=int)
df
```

```python
X=df.drop('Survived',axis='columns')
Y = df.Survived
```
