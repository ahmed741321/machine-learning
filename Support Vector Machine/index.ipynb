```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#sow chart data

plt.scatter(iris.data[:,2],iris.data[:,3],c=iris.target)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.sample(9)
```

```python
iris.target_names
```

```python
df['target']=iris.target
df.sample(9)
```

```python
df.drop('target',axis='columns',inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(df,iris.target,test_size=0.2)

```

```python
from sklearn.svm import SVC

model=SVC(kernel='rbf',C=30,gamma='auto')
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
# Plot 1: Sepal Length vs Petal Length
plt.figure(figsize=(10, 5))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], c=iris.target, cmap='viridis', edgecolor='k')
plt.title('Sepal Length vs Petal Length')

# Plot 2: Model Predictions vs Petal Length
predictions = model.predict(X_test)
plt.figure(figsize=(10, 5))
plt.scatter(predictions, X_test.iloc[:, 2], c=Y_test, cmap='viridis', edgecolor='k')  # Accessing column by position
plt.title('Model Predictions vs Petal Length')

# Show the plots
plt.show()
```
