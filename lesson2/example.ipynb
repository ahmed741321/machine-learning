```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("C:/Users/ahahm/Desktop/machine learning/lesson2/example_treain.csv")
df.sample(9)
```

```python
df.info()
```

```python
new_df=df[["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd","SalePrice"]]
new_df.sample(9)
new_df.reset_index(drop=True)
```

```python
reg=linear_model.LinearRegression()
reg.fit(new_df[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']],new_df['SalePrice'])
```

```python
data_new_predict=new_df.copy()
data_new_predict
```

```python
data_new_predict["new price"]=reg.predict(data_new_predict[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']])
data_new_predict.sample(9)
```

```python
# رسم النقاط الأصلية
plt.scatter(data_new_predict["YearBuilt"], data_new_predict["SalePrice"], color="red", label="Original Data", marker="+")

# رسم نقاط التوقعات
plt.scatter(data_new_predict["YearBuilt"], data_new_predict["new price"], color="green", label="Predicted Data", marker=".")

# إضافة العناوين والتسميات
plt.title("Linear Regression: Area vs Price")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.legend()

# عرض الرسم
plt.show()
```
